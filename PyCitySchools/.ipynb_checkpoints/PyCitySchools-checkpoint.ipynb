```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
bins = [0, 69, float('inf')]
group_names=['fail','pass']
school_data_complete['overall_score']=(school_data_complete['math_score']+school_data_complete['reading_score'])/2
school_data_complete['math pass'] = pd.cut(school_data_complete['math_score'], bins, labels=group_names, include_lowest=True)
school_data_complete['reading pass'] = pd.cut(school_data_complete['reading_score'], bins, labels=group_names, include_lowest=True)
school_data_complete['overall pass'] = pd.cut(school_data_complete['overall_score'], bins, labels=group_names, include_lowest=True)
Dist_df=pd.DataFrame(
    {'Total Schools': [school_data['School ID'].count()],
     'Total Students': [student_data['Student ID'].count()],
     'Total Budget': [school_data['budget'].sum()],
     'Average Math Score':[school_data_complete['math_score'].mean()],
     'Average Reading Score':[school_data_complete['reading_score'].mean()],
    'Passing Math': [school_data_complete['math pass'].value_counts()['pass']/student_data['Student ID'].count()*100],
    'Passing Reading': [school_data_complete['reading pass'].value_counts()['pass']/student_data['Student ID'].count()*100]}
)

Dist_df['Overall Passing Rate'] = (Dist_df['Passing Math']+Dist_df['Passing Reading'])/2
Dist_df['Passing Math'] = Dist_df['Passing Math'].astype(float).map("{0:.2f}%".format)
Dist_df['Passing Reading'] = Dist_df['Passing Reading'].astype(float).map("{0:.2f}%".format)
Dist_df['Overall Passing Rate'] = Dist_df['Overall Passing Rate'].astype(float).map("{0:.2f}%".format)
Dist_df['Total Budget'] = Dist_df['Total Budget'].astype(float).map("${:,.2f}".format)
Dist_df
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
schools=school_data_complete.groupby(['school_name'])

school_df=pd.DataFrame({
    'School Type': schools['type'].first(),
    'Total Students':schools.size(),
    'Total School Budget':schools['budget'].first(),
    'Per Student Budget': schools['budget'].first()/schools.size(),
    'Average Math Score': schools['math_score'].mean(),
    'Average Reading Score': schools['reading_score'].mean(),
    'Passing Math': (school_data_complete[school_data_complete['math_score']>=70].groupby(['school_name']).size()/schools.size())*100,
    'Passing Reading': (school_data_complete[school_data_complete['reading_score']>=70].groupby(['school_name']).size()/schools.size())*100,}
)

school_df['Overall Passing Rate'] = (school_df['Passing Math']+school_df['Passing Reading'])/2
unformatedschool_df=school_df.copy()
school_df['Passing Math'] = school_df['Passing Math'].astype(float).map("{0:.2f}%".format)
school_df['Passing Reading'] = school_df['Passing Reading'].astype(float).map("{0:.2f}%".format)
school_df['Overall Passing Rate'] = school_df['Overall Passing Rate'].astype(float).map("{0:.2f}%".format)
school_df=school_df.sort_values(by=['Overall Passing Rate'], ascending=False)
school_df.index.name=''
top_schools=school_df.head()
top_schools
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
school_df=school_df.sort_values(by=['Overall Passing Rate'])
bot_schools=school_df.head()
bot_schools
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
spending = unformatedschool_df.loc[:,['Average Math Score',
                                  'Average Reading Score','Passing Math',
                                  'Passing Reading','Overall Passing Rate',]]
spending['Spending Ranges (Per Student)']= pd.cut(unformatedschool_df['Per Student Budget'],spending_bins,labels=group_names)
spending = spending.groupby('Spending Ranges (Per Student)').mean()
spending['Passing Math'] = spending['Passing Math'].astype(float).map("{0:.2f}%".format)
spending['Passing Reading'] = spending['Passing Reading'].astype(float).map("{0:.2f}%".format)
spending['Overall Passing Rate'] = spending['Overall Passing Rate'].astype(float).map("{0:.2f}%".format)
spending.index.name = None
spending
```
